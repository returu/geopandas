```python
import geopandas as gpd
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world = world[world.name!="Antarctica"]
world
```

```python
world.plot()
```

```python
# 选择前10行数据
gdf = world[:10]
gdf
```

```python
# 绘制地理空间图
gdf.plot(kind="geo")
```

```python
# 散点图
gdf.plot(kind="scatter" , x = "iso_a3" , y = "gdp_md_est")
```

```python
# 折线图
gdf.plot(kind="line" ,  y="gdp_md_est" ,marker="*" , markersize=10)
```

### 还可以使用 GeoDataFrame.plot.kind 访问器方法创建这些其他图，而不需提供 kind 关键字参数。

```python
gdf.plot.bar()
```
