## GeoDataFrame.explore()：

```python
gdf.explore()
```

```python
gdf.crs
```

## 参数设置：

```python
gdf.explore(column="NAME", # 用于绘制分级统计图choropleth的列
           tooltip="PAC", # 当鼠标悬停时显示的列的值，例如：PAC 110106
           popup = True, # 当为True时，鼠标点击几何对象时会显示全部列的信息
           tiles = "CartoDB positron", # 设置地图底图，默认为OpenStreetMap
           cmap = "tab20", # 定义颜色
           style_kwds = {"color":"black"}, # 设置黑色边框
           )
```

## folium.Map对象：

```python
# 加载点数据
points = gpd.read_file("./datasets/china/china_counties_points.shp")
points = points[:30]
points
```

```python
import folium

m = gdf.explore(
     column="Shape_Area",  # make choropleth based on "BoroName" column
     scheme="naturalbreaks",  # 自然间断点
     k=10, # 划分为 10 类
     legend_kwds=dict(colorbar=False), # 不使用colorbar
     name="行政区划" # 地图中的图层名称
)

points.explore(m=m,# 直接传递folium.Map对象
              color="red",
              marker_kwds={"fill":True , "radius":10 , "size":5},# 设置点的显示样式
              tooltip="NAME",# 当鼠标悬停时显示的列的值
              tooltip_kwds=dict(labels=False),# 当鼠标悬停时不显示列名
              name="points" # 地图中的图层名称
              )

# 指定地图底图
folium.TileLayer("Stamen Terrain" , control=True).add_to(m)
# 添加图层控件，以便在地图中切换图层
folium.LayerControl().add_to(m)

m
```
